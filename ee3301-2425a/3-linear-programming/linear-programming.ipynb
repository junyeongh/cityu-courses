# 3. Linear Programming

## Example 2

p. 22

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x
x = np.linspace(-10, 10, 400)

# Define the equations
y1 = 14 - 2*x
y2 = (8 + x) / 2
y3 = 2*x - 10

# Plot the equations
plt.plot(x, y1, label='2x + y = 14')
plt.plot(x, y2, label='-x + 2y = 8')
plt.plot(x, y3, label='2x - y = 10')

# Set limits to focus on the first quadrant
plt.xlim(0, 10)
plt.ylim(0, 10)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Linear Equations')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
